# Basic Graphics

This notebook shows how to create simple graphics from your data. It is based on the [DataCamp MatPlotLib tutorial](https://campus.datacamp.com/courses/intermediate-python-for-data-science/) and the [MatPlotLib](https://matplotlib.org/) tutorials.

### Setting things up

First you have to *import* the MatPlotLib library and you have to set up the notebook so the graphics are rendered into the notebook.

```python
import matplotlib.pyplot as plt
import numpy
%matplotlib inline
```

### Simple plot

Here is a simple plot.

```python
points = [1,2,4,3] # Set up a list of data points.
plt.plot(points) # Describe the plot.
plt.show() # Show the plot.
```

Now we can add some labels and control the scale and ticks on the axes.

```python
plt.plot(points) # Describe the plot.

plt.xlabel('X Axis') # X Axis Label
plt.ylabel('Y Axis') # Y Axis Label
plt.title('Plot Title') # Plot Title
plt.xticks([0,1,2,3], ['1','2','3','4']) # Relabeling the X-Axis ticks
plt.yticks([0,1,2,3,4,5]) # Setting the Y-Axis scale

plt.show() # Show the plot.
```

### Importing some data

Now we import some data from a CSV into three lists to make some more realistic plots.

```python
import csv

dateList = [] # We create 3 lists for the data
analytic = []
clout = []

with open('LIWCdatashort.csv', 'r') as file: # This makes sure that file is closed after reading
    data = csv.reader(file)
    for row in data:
        dateList.append(row[0]) # This puts all the first column into a list called "dateList"
        analytic.append(row[1]) # This puts all the first column into a list called "analytic"
        clout.append(row[2]) # This puts all the first column into a list called "clout"
file.closed

# Remove the first item from each list as it is the label for the list.
del dateList[0]
del analytic[0]
del clout[0]

# Conver items in numeric lists into floating point numbers
analyticFloat = [float(item) for item in analytic]
cloutFloat = [float(item) for item in clout]

cloutFloat
```

### Double plot

Here is a plot of two sets of values from the CSV.

```python
plt.plot(analyticFloat)
plt.plot(cloutFloat)

plt.show()
```

Here is a version with some labels and stuff.  Note how we set the X-axis ticks to be the dates from the CSV. Note also how we set the size of the *figure* in which we have the plot.

```python
plt.figure(figsize=(10,5)) # This sets the size of the figure.
plt.plot(analyticFloat)
plt.plot(cloutFloat)
plt.xticks(range(0,len(analyticFloat)),dateList,rotation=70) # This sets the xticks to labes from the list dateList
plt.grid(True) # This adds a grid
plt.show()
```

### Create a histogram

We can also create other types of graphics like histograms. A histgram shows you how many items fit in a set of bins.

```python
plt.hist(analyticFloat)

plt.show()
```

You get a very different histogram if you change the number of bins.

```python
plt.hist(analyticFloat, bins=5)

plt.show()
```

### Scatter Plot

You can also create scatter plots if you have more than one list.

```python
plt.scatter(analyticFloat,cloutFloat)

# Customizations
plt.xlabel('Analytic')
plt.ylabel('Clout')

plt.show()
```

Here is a more complex *scatter plot* that has labels next to the points and a text annotation.

```python
# Scatter plot
plt.scatter(x = analyticFloat, y = cloutFloat, alpha = 0.8)

# Customizations
plt.xlabel('Analytic')
plt.ylabel('Clout')
plt.title('Months of Data')

for label, x, y in zip(dateList, analyticFloat, cloutFloat):
    plt.annotate(
        label, 
        xy = (x, y), 
        xytext = (-5, 5),
        textcoords = 'offset points', 
        ha = 'left', 
        va = 'bottom',
        )


# Additional annotation
plt.text(67, 85.75, 'Note that the months start in upper left.')

# Add grid
plt.grid(True)

# Show the plot
plt.show()
```

### Combining plots in a single figure

You can combine more than one plot in a single figure in order to compare them.

```python
plt.figure(figsize=(10,7))

# This sets the subplot to have 2 rows and 1 column of which this subplot is number 1
plt.subplot(211) 
plt.plot(cloutFloat)
plt.ylabel('Clout')
plt.xlabel('Clout Trend')

# This sets the subplot to have 2 rows and 1 column of which this subplot is number 2
plt.subplot(212) 
plt.plot(analyticFloat)
plt.ylabel('Analytic')
plt.xlabel('Analytic Trend')

plt.show()
```

### Saving a plot

You can save your plots to a file.

```python
plt.plot(cloutFloat)
plt.xlabel('Clout')

plt.savefig("Clout.png") # The extension controls the type of image. Eg. use "foo.pdf" to save as PDF.
print("Done")
```
